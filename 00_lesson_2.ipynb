```python
import os
import google.generativeai as genai
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

# Configure the API key
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
```

```python
# Initialize the model
model = genai.GenerativeModel('gemini-2.5-flash')
```

```python
# Make your first API call
response = model.generate_content(
    "Hello! What can you do?",
    generation_config=genai.GenerationConfig(
        temperature=0.7,    # control randomness in responses
                            # 0-0.3: Factual, consistent responses
                            # 0.4-0.7: Balanced creativity
                            # 0.8-1.0: Creative, varied responses
        top_p=0.1,  # considers tokens with top cumulative probability mass
                    # 0.1: Very focused, predictable
                    # 0.5: Moderately diverse
                    # 1.0: Consider all options
    )
)
```

```python
response
```

```python
# Print the response
print(response.text)
```

# Example: Different Parameter Settings

```python
# Creative writing
response = model.generate_content(
    "Write a story about a dog",
    generation_config=genai.GenerationConfig(
        temperature=0.9,
        max_output_tokens=500,
        top_p=0.9
    )
)
# Print the response
print(response.text)
```

```python
# Factual Q&A
response = model.generate_content(
    "What is the capital of Vietnam?",
    generation_config=genai.GenerationConfig(
        temperature=0.1,
        max_output_tokens=50,
        top_p=0.1
    )
)
# Print the response
print(response.text)
```

```python
# Code generation
response = model.generate_content(
    "Write a Python function that takes a string and returns an integer",
    generation_config=genai.GenerationConfig(
        temperature=0.3,
        max_output_tokens=300,
    )
)
# Print the response
print(response.text)
```
