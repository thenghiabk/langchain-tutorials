````python
import os
import google.generativeai as genai
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

# Configure the API key
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
````

````python
# Initialize the model
model = genai.GenerativeModel('gemini-2.5-flash')
````

# Role Playing

````python
PROMPT = """
You are an experienced Python developer and code reviewer.
Your task is to review the following code and provide:
1. Potential bugs or issues
2. Performance improvements
3. Best practices violations
4. Security concerns

Code to review:
```python
items = {"Coffee": 2.2, "Tea": 1.5, "Chocolate": 2.5}

for item in items.keys()
     income = 0
      qty = input(f"How many {item}s have you sold? ")
     income = income + qty * items[item]
print(f"\nThe income today was £{income:0.2f}")
```
"""
````

````python
# Make your first API call
response = model.generate_content(
    contents=PROMPT,
    generation_config=genai.GenerationConfig(
        temperature=0.7,    # control randomness in responses
                            # 0-0.3: Factual, consistent responses
                            # 0.4-0.7: Balanced creativity
                            # 0.8-1.0: Creative, varied responses
        top_p=0.1,  # considers tokens with top cumulative probability mass
                    # 0.1: Very focused, predictable
                    # 0.5: Moderately diverse
                    # 1.0: Consider all options
    )
)
# Print the response
print(response.text)
````
